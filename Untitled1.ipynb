## Krebstexte

Für dieses Projekt habe ich ungefähr tausend Threads von  cancerforums.net/threads/ heruntergeladen. Schauen wir uns ersteinmal die Texte an:

```python
import pandas as pd
pd.set_option('display.max_rows', 1000)
import numpy as np
import csv

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('poster')
sns.set_style('white')
#sns.set_color_codes()
plot_kwds = {'alpha' : 0.35, 's' : 30, 'linewidths':0}
```

```python
df = pd.read_csv('cancer.csv', encoding = "ISO-8859-1")
dl = df['text'].tolist()
mytext = " ".join(str(x) for x in dl)
print(mytext)
```

Viele Krebstexte. Was sind denn die häufigsten Worte in diesen Texten? (Leo auf Platz 12 ist Urheber der Seite...) Platz 25&26.. The doctor said...

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist



ps = PorterStemmer()
stopWords = set(stopwords.words('english'))
content = pd.DataFrame([])

#.replace('|',' ')

stemtext = ps.stem(mytext)
tokens = nltk.word_tokenize(mytext)
tokens = [word.lower() for word in tokens if word.isalpha()]

tokensf = []

for w in tokens:
    if w not in stopWords:
        tokensf.append(w)

frequencies = FreqDist(tokensf).most_common(100)
frequenciesdf = pd.DataFrame(frequencies, columns=['word','frequency'])
print(frequenciesdf)
```

Jetzt schauen wir uns an wie die Texte sich natürlich gruppieren...

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import re
vec = TfidfVectorizer()

#vec = TfidfVectorizer(token_pattern=r'(?<=[^|;])[\s\w]+(?=[$|;])')
X = vec.fit_transform(dl)
#print(pd.DataFrame(X.toarray(), columns=vec.get_feature_names()))
```

```python
Xfr = pd.DataFrame(X.toarray())
from sklearn.decomposition import PCA

PCA =  PCA(n_components=30) 
XPCA = PCA.fit_transform(Xfr)
dPCA = pd.DataFrame(XPCA)
```

Das ist noch nicht so übersichtlich...

```python
plt.scatter(dPCA[0], dPCA[1], color='b', **plot_kwds)
```

```python
from sklearn.metrics.pairwise import pairwise_distances

XCOS = pairwise_distances(XPCA, metric='cosine')
```

```python
from sklearn.manifold import TSNE


TSNE = TSNE(n_components=2, metric='precomputed', perplexity=36)#random_state=10
XTSNE = TSNE.fit_transform(XCOS)
dTSNE = pd.DataFrame(XTSNE)
plt.scatter(dTSNE[0], dTSNE[1], color='grey', **plot_kwds)
```

```python
import hdbscan

clusterer = hdbscan.HDBSCAN(min_cluster_size=26, gen_min_span_tree=True)
clusterer.fit(XTSNE)
XCLUST = clusterer.labels_


clusternum = len(set( clusterer.labels_))-1
#clusterer.condensed_tree_.plot()

color_palette = sns.color_palette('deep', 160)
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]

plt.scatter(dTSNE[0], dTSNE[1], s=30, linewidth=0, c=cluster_colors, alpha=0.25)
dfclust = pd.DataFrame(XCLUST)



dfclust = pd.concat([df,dfclust],axis=1)
dfclust.columns = ['text','cluster']

drc = pd.concat([dTSNE[0], dTSNE[1],dfclust],axis=1)
drc.columns = ['xTSNE','yTSNE','text','cluster']
```

Das ist schon viel besser. Jetzt müssen wir nur noch Wissen welche Gruppen was enthalten. Ich meine, dass man Krebsarten ganz gut unterscheiden kann, z. B. prostate und husband.

```python
content = pd.DataFrame([])
for x in range(0,clusternum):
    subtext = list(dfclust.loc[dfclust['cluster'] == x]['text'])
    subtext = " ".join(str(x) for x in subtext).replace('|',' ').replace('\\t',' ').replace('\\n',' ')

    subtext = ps.stem(subtext)
    tokens = nltk.word_tokenize(subtext)
    tokens = [word.lower() for word in tokens if word.isalpha()]

    tokensf = []

    for w in tokens:
        if w not in stopWords:
            tokensf.append(w)

    content[x] = FreqDist(tokensf).most_common(60)
    
    
```

```python
flatui = [ "#95a5a6","#3366cc", "#dc3912", "#ff9900", "#109618", "#990099", "#0099c6", "#dd4477", "#66aa00", "#b82e2e", "#316395", "#994499", "#22aa99", "#aaaa11", "#6633cc", "#e67300", "#8b0707", "#651067", "#329262", "#5574a6", "#3b3eac"]

sns.set_palette(sns.color_palette(flatui))
sns.lmplot(data=drc, x='xTSNE', y='yTSNE', hue='cluster', scatter_kws={"s": 20},
                   fit_reg=False, legend=False)
lgnd = plt.legend(loc=(1.04,0), scatterpoints=1, fontsize=15)
for x in range(0, len(lgnd.legendHandles)):
    lgnd.legendHandles[x]._sizes = [70]


print(content)
```
